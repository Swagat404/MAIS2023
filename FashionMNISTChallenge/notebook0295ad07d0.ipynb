```python
train_images = np.load("/kaggle/input/mais-202-winter-2023/train_images.npy")
print(train_images.shape)

import matplotlib.pyplot as plt

def show_image(arr):
    two_d = (np.reshape(arr, (28, 28)) * 255).astype(np.uint8)
    plt.imshow(two_d, interpolation='nearest')
    plt.show()
```

```python
test_images = np.load("/kaggle/input/mais-202-winter-2023/test_images.npy")
```

```python
test_images.shape
```

```python
train_images.shape
```

```python
n_features = 28*28
data_reshaped_test = test_images.reshape(20000, n_features)
```

```python
n_features = 28*28
data_reshaped_train = train_images.reshape(50000, n_features)
```

```python
data_reshaped_train
```

```python
df = pd.read_csv("/kaggle/input/mais-202-winter-2023/train_labels.csv")
labels=df['label']
print(labels.shape)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Initialize the GradientBoostingClassifier
clf = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3)

# Fit the model on the training data
clf.fit(data_reshaped_train, labels)
```

```python
y_pred = clf.predict(data_reshaped_test)
```

```python
import pickle
pickle.dump(clf, open("model1", 'wb'))
```

```python
print(y_pred)
```

```python
df_test = pd.read_csv('/kaggle/input/mais-202-winter-2023/sample_submission.csv')
df_test['label'] = y_pred
df_test.to_csv('submission.csv', index=False)
```
